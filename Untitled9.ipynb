```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data into a Pandas DataFrame
df = pd.read_csv('cencus_data.csv')

# Histogram for Age
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=10, edgecolor='k')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()

# Scatter Plot for Education Level vs. Income
plt.figure(figsize=(8, 6))
plt.scatter(df['Education'], df['Income'], alpha=0.5)
plt.xlabel('Education Level')
plt.ylabel('Income')
plt.title('Education Level vs. Income')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

# Bar Chart for Marital Status vs. Average Income
marital_status_avg_income = df.groupby('Marital_Status')['Income'].mean()
plt.figure(figsize=(8, 6))
marital_status_avg_income.plot(kind='bar', color='skyblue')
plt.xlabel('Marital Status')
plt.ylabel('Average Income')
plt.title('Marital Status vs. Average Income')
plt.xticks(rotation=45)
plt.show()
```
